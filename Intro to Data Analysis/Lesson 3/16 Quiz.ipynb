```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

filename = 'nyc_subway_weather.csv'
subway_df = pd.read_csv(filename)

### Write code here to group the subway data by a variable of your choice, then
### either print out the mean ridership within each group or create a plot.
```

```python
subway_df_weekday = subway_df[subway_df['weekday'] == 1]
```

```python
subway_df_weekday_rain = subway_df_weekday[subway_df_weekday['rain'] == 1]
subway_df_weekday_no_rain = subway_df_weekday[subway_df_weekday['rain'] == 0]
```

```python
entries_rain = subway_df_weekday_rain.groupby('TIMEn').mean()['ENTRIESn_hourly']
exits_rain = subway_df_weekday_rain.groupby('TIMEn').mean()['EXITSn_hourly']
print entries_rain, '\n'*2, exits_rain
```

```python
entries_no_rain = subway_df_weekday_no_rain.groupby('TIMEn').mean()['ENTRIESn_hourly']
exits_no_rain = subway_df_weekday_no_rain.groupby('TIMEn').mean()['EXITSn_hourly']
print entries_no_rain, '\n'*2, exits_no_rain
```

```python
entries_rain.plot(label='Entries Rain')
entries_no_rain.plot(label='Entries No Rain')
plt.title('Entries on Rainy and Non-rainy Days')
plt.xlabel('Time of Day')
plt.ylabel('Entries')
plt.legend(loc='lower right')
plt.show()
```

```python
exits_rain.plot(label='Exits Rain')
exits_no_rain.plot(label='Exits No Rain')
plt.title('Exits on Rainy and Non-rainy Days')
plt.xlabel('Time of Day')
plt.ylabel('Exits')
plt.legend(loc='lower right')
plt.show()
```
